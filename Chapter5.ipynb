```python
import pandas as pd
import numpy as np
```

```python
# Example 5-1
data = [0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.60, 0.70]
RelPermKrow = pd.DataFrame(data, columns = ['Sw'])
RelPermKrowPirson = RelPermKrow.copy()
RelPermKrowPirson
```

```python
Swc = 0.25 # Connate Water Saturation
RelPermKrow['Swx'] = round((RelPermKrow['Sw']-Swc)/(1-Swc),4) # Equation 5-2
RelPermKrow['Kro'] = round((1-RelPermKrow['Swx'])**3,3) # Equation 5-4 Drainage Oil-Water Relative Permeabilities
RelPermKrow['Krw'] = round(RelPermKrow['Swx']**3,4) # Equation 5-4
RelPermKrow
```

```python
data = [0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70]
RelPermKrgo = pd.DataFrame(data, columns = ['Sg'])
RelPermKrgo
RelPermKrgoCorey = RelPermKrgo.copy()
```

```python
RelPermKrgo['So'] = 1 - RelPermKrgo['Sg'] - Swc # Equation 5-2
RelPermKrgo['Sox'] = round((RelPermKrgo['So'])/(1-Swc),3) # Equation 5-2
RelPermKrgo['Kro'] = round((RelPermKrgo['Sox'])**3,3) # Equation 5-7 Drainage Gas-Oil Relative Permeabilities
RelPermKrgo['Krg'] = round((1-RelPermKrgo['Sox'])**3,3) # Equation 5-7
RelPermKrgo
```

```python
# Example 5-2: Pirson's Correlation
RelPermKrowPirson['Swx'] = round((RelPermKrowPirson['Sw']-Swc)/(1-Swc),4) # Equation 5-2
RelPermKrowPirson['Krwp'] = round(np.sqrt(RelPermKrowPirson['Swx']) * RelPermKrowPirson['Sw'] ** 3, 3)
RelPermKrowPirson['Kro'] = round((1-RelPermKrowPirson['Swx'])*(1-RelPermKrowPirson['Swx']**0.25*np.sqrt(RelPermKrowPirson['Sw']))**0.5,3)
RelPermKrowPirson
```

```python
# Example 5-3: Corey's Method
RelPermKrgoCorey['Sgx'] = round((RelPermKrgoCorey['Sg'])/(1-Swc),3) # Equation 5-2
RelPermKrgoCorey['Kro'] = round(RelPermKrgoCorey['Sgx']**4,3) # Equation 5-7 Drainage Gas-Oil Relative Permeabilities
RelPermKrgoCorey['Krg'] = round(RelPermKrgoCorey['Sgx']**3*(2-RelPermKrgoCorey['Sgx']),3) # Equation 5-7
RelPermKrgoCorey
```
